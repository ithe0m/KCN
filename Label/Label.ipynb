# Delete The Identical Data Rows:

```python
similarity_Matrix=np.zeros((len(DATA),len(DATA)))
for i in range(len(DATA)):
  for j in range(len(DATA)):
    if(i!=j):
      if(DATA.iloc[i,0]==DATA.iloc[j,0])and(DATA.iloc[i,1]==DATA.iloc[j,1])and(DATA.iloc[i,2]==DATA.iloc[j,2])and(DATA.iloc[i,3]==DATA.iloc[j,3])and(DATA.iloc[i,4]==DATA.iloc[j,4]):
        similarity_Matrix[i,j]=1
```

```python
DATA_F = DATA.copy()
index=[]
for i in range(len(DATA)):
  index.append(str(i))
DATA_F.index=index

drop_list=[]
for i in range(len(DATA)):
  for j in range(i,len(DATA)):
    if(similarity_Matrix[i,j]==1):
      if(DATA_F.index[j] not in drop_list):
        drop_list.append(DATA_F.index[j])

DATA_F.drop(drop_list,axis=0,inplace=True)
```
