**Polar Map Function Defined:**

```python
path = '/content/drive/MyDrive/KCN/Sample/'
def polar_map(file_path):
  heat_map = pd.read_csv(file_path,sep=';')
  heat_map.iloc[0,1]=0
  value = np.zeros(len(heat_map.iloc[:,3])-1)
  z= np.zeros((360,51))
  z[:,0]=heat_map.iloc[0,3]
  k = 1
  m = 0
  for i in range(1,int((len(heat_map)-1)/2+1)):
    value[i-1]= heat_map.iloc[k,3]
    k=k+1
    m=m+1
    if(m==50):
      k = k+50
      m=0

  k = 51
  m = 0
  for i in range(int((len(heat_map)-1)/2),int((len(heat_map)-1))):
    value[i]= heat_map.iloc[k,3]
    k=k+1
    m=m+1
    if(m==50):
      k = k+50
      m=0

  k=0
  for i in range(360):
    for j in range(1,51):
      z[i,j] = value[k]
      k=k+1

  rad = np.linspace(0, 5, 51)
  degree = np.linspace(0, (2 * np.pi), 360)
  r, th = np.meshgrid(rad, degree)

  return degree,th,r,z


```

**PLOT**

```python
degree,th,r,z = polar_map(path + '._gholam_rezai_kashani.narges_OD_2019_09_17.01_01_39_484_Post. Elevation BFTA.csv')

cmap = ListedColormap(["#C652F8", "#B83EE8", "#B129D1", "#9B29B9","#832DA5","#672BA7","#50298C","#2C0098","#3000C6","#2737E5","#2867E3","#1B99FA","#03A6F8","#00C2FC","#04D4F6","#00F9FE","#14DA52","#5EEB06","#8FFD06","#FBFF05","#FFEB00","#E1D303","#DEAD00","#E69001","#E56F00","#E25202","#CD3608","#C20506","#A30001","#943C41","#AF5454","#CE6767","#E37A7A"])

fig = plt.figure(figsize=(8,8), dpi=100)
plt.subplot(projection="polar")
plt.style.use('classic')
plt.pcolormesh(th, r, z,cmap = cmap )
plt.plot(degree, r, color='k', ls='none') 
plt.grid()
plt.colorbar().set_label('[μm]')
plt.title(label='Post. Elevation BFTA [μm]  (OD)',loc='left',backgroundcolor='gray',pad='40')
plt.clim(-85, 75);
plt.ylim([0,4.5])

plt.savefig(path +'Pic/._gholam_rezai_kashani.narges_OD_Post. Elevation BFTA.jpeg')

plt.show()
```
